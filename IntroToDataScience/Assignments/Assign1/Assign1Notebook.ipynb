```python
df = pd.read_csv("IndiaFiltered.csv")
commercial_expo = df["CommercialExportUSD"].tolist()
commercial_expo = commercial_expo[15:]
commercial_expo = np.array(commercial_expo)
print("Commercial Export in USD\n",commercial_expo)
```

```python
GDI = df["GDI"].tolist()
GDI = GDI[15:]
GDI = np.array(GDI)
print("Gross Domestic Income in INR\n",GDI)
```

```python
plt.boxplot(commercial_expo)
plt.title("Commercial Export in USD 1975-2024")
```

```python
"""
The second interquartile range massively outperforms the first interquartile range.
This indicates that half of the data is above the median value showing a huge skewness
"""
```

```python
plt.boxplot(GDI)
plt.title("Gross Domestic Income in Rupees 1975-2024")
```

```python
"""
Maximum value is very far away from third quartile, indicating income disparities
"""
```

```python
years = np.array(df["Year"])
years = years[15:]
plt.figure()
plt.xlabel("GDI")
plt.ylabel("Annual Commercial Export in USD")
plt.title("GDI vs Annual Commercial Export(USD)")
plt.plot(GDI, commercial_expo)
```
