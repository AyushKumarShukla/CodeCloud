```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import add_dummy_feature
```

```python
data_root = "https://github.com/ageron/data/raw/main/"
sat = pd.read_csv(data_root + "lifesat/lifesat.csv")

gdp = sat[["GDP per capita (USD)"]].values
countrysat = sat[["Life satisfaction"]].values

sat.plot(kind="scatter", grid=True, x="GDP per capita (USD)", y="Life satisfaction");
plt.axis([23_500,62_500,4,9])

model = LinearRegression()

model.fit(gdp,countrysat)

new_values = np.linspace(gdp.min(),gdp.max(),100).reshape(-1,1)
predictions = model.predict(new_values)

plt.plot(new_values,predictions,color="blue",linewidth=2, label="Predictions")
plt.show()
```

```python
#set seed for generating probability distros
np.random.seed(42)
m = 100
#1...generating training dataset
#generate a column vector via normal distro
X = 2 * np.random.rand(m,1)
#use the above column vector and add Gaussian noise to it
y = 4 + 3*X + np.random.randn(m,1)

#2...training using closed form solution
# add a dummy feature to X
X_dummy = add_dummy_feature(X)

#apply normal equation
theta_best = np.linalg.inv(X_dummy.T @ X_dummy) @ X_dummy.T @ y

#3...making predictions
X_new = np.array([[0],[2]])
X_new_dummy = add_dummy_feature(X_new)
y_predict = X_new_dummy @ theta_best
print(y_predict)

plt.plot(X_new,y_predict)
plt.plot(X,y,"b.")
plt.show()
```
